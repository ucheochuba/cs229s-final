```python
# Compute percentiles at intervals of 10
percentiles = np.percentile(filtered_scores, np.arange(10, 100, 2))

# Display the percentiles
for i, p in zip(range(10, 100, 2), percentiles):
    print(f"{i}th Percentile: {p:.2e}")
```

```python
percentiles = np.percentile(filtered_scores, np.arange(0, 101, 1))  # 0th to 100th percentiles
cdf = np.linspace(0, 100, len(percentiles))

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(cdf, percentiles, marker='o', color='b', linestyle='-', markersize=5)
plt.title('Cumulative Percentile Plot of Scores')
plt.xlabel('Percentile')
plt.ylabel('Score Value')
plt.grid(True)
plt.show()
```

```python
# Assuming filtered_scores is available (already cleaned)
# Sort the scores to compute the CDF
sorted_scores = np.sort(filtered_scores)

# Calculate the cumulative proportion of the data
cdf_values = np.arange(1, len(sorted_scores) + 1) / len(sorted_scores)

# Compute percentiles (0th to 100th) and corresponding values
percentiles = np.percentile(filtered_scores, np.arange(0, 101, 2))

# Compute the CDF values for the percentiles
cdf_percentiles = np.array([np.sum(sorted_scores <= p) / len(sorted_scores) for p in percentiles])

# Plot the CDF
plt.figure(figsize=(8, 6))
plt.plot(np.arange(0, 101, 2), cdf_percentiles, marker='o', color='b', linestyle='-', markersize=5)
plt.title('Cumulative Distribution Function (CDF) of Scores by Percentiles')
plt.xlabel('Percentile')
plt.ylabel('Cumulative Proportion')
plt.ylim(0, 1) 
plt.grid(True)
plt.show()
```
