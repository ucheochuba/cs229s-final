```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
attn_scores = np.load('scores_tensor.npy')
print("Loaded scores shape:", attn_scores.shape)
```

```python
print(attn_scores)
```

```python
scores_flat = attn_scores.flatten()
```

```python
plt.hist(scores_flat, bins=10, color='blue', alpha=0.7)  # Adjust bins as needed
plt.title('Histogram of Scores Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
filtered_scores = scores_flat[np.isfinite(scores_flat)]

# Compute summary statistics
mean = np.mean(filtered_scores)
median = np.median(filtered_scores)
std_dev = np.std(filtered_scores)
min_val = np.min(filtered_scores)
max_val = np.max(filtered_scores)
percentiles = np.percentile(filtered_scores, [10, 20, 30, 40, ])  # 90th, 95th, 99th percentiles

# Display results
print("Summary Statistics:")
print(f"Mean: {mean:.2e}")
print(f"Median: {median:.2e}")
print(f"Standard Deviation: {std_dev:.2e}")
print(f"Minimum Value: {min_val:.2e}")
print(f"Maximum Value: {max_val:.2e}")
print(f"90th Percentile: {percentiles[0]:.2e}")
print(f"95th Percentile: {percentiles[1]:.2e}")
print(f"99th Percentile: {percentiles[2]:.2e}")
```

```python
# Filter scores between a range of interest
filtered_scores = scores_flat[np.isfinite(scores_flat)]


plt.hist(filtered_scores, bins=10, color='blue', alpha=0.7)  # Adjust bins as needed
plt.title('Histogram of Scores Values')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# Compute percentiles at intervals of 10
percentiles = np.percentile(filtered_scores, np.arange(10, 100, 10))

# Display the percentiles
for i, p in zip(range(10, 100, 10), percentiles):
    print(f"{i}th Percentile: {p:.2e}")
```
